```python
# Make a simple calculator that can add, subtract, multiply and divide two  number using functions
def add(a, b):
  s = a +b
  return s

def sub(a, b):
  s = a -b
  return s

def multi(a, b):
  s = a *b
  return s

def div(a, b):
  s = a /b
  return s

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

choice = 0
while choice != 5:
  print("\n2what operation do you want to do:")
  print("1.Addition")
  print("2.Subtraction")
  print("3.Multiplication")
  print("4.Division")
  print("5.Exit")
  choice = int(input("Enter your choice: "))
  if choice == 1:
   print(num1,"+",num2,"=",add(num1,num2))
  elif choice == 2:
    print(num1,"-",num2,"=",sub(num1,num2))
  elif choice == 3:
    print(num1,"*",num2,"=",multi(num1,num2))
  elif choice == 4:
    print(num1,"/",num2,"=",div(num1,num2))
  elif choice == 5:
    print("Thank you")
  else:
    print("Invalid choice")
```

## Exercise on List and Dictionaries

```python
items = [1,2,2,3,3,4,5]

seen_items = []

for i in range(len(items)):
    current_number = items[i]
    if current_number in seen_items:
        print(f"Found the first duplicate: {current_number}")
        items.pop(i)
        break
    else:
        seen_items.append(current_number)

print("Final list:", items)
```

```python
sample_dict = {
    "name": "Anderson",
    "age": 35,
    "salary": 5000,
    "city": "London"
}
keys = ["name", "salary"]
new_dict = {}
for key in keys:
    new_dict[key] = sample_dict[key]

print(new_dict)
```

```python
list_one = ['one', 'two', 'three']
list_two = [1, 2, 3]
result_dict = {}
for i in range(len(list_one)):
    key = list_one[i]
    value = list_two[i]
    result_dict[key] = value

print(result_dict)
```
